```python
#---------------------------------------------------
# Import Packages

import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('college.csv')
```

```python
#Show data frome head
df.head()
```

```python
#
df.describe()
```

```python
# plot a trellis grid

sns.set(style = 'white')

g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, lw=3)
```
